# **Tìm hiểu SimpleImpute, sử dụng strategy là mean**




```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
data1 = pd.read_csv("categorical_data.csv")
```

```python
data1
```

Thay thế các giá trị còn thiếu bằng cách điền các giá trị trung bình

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
data1.iloc[:, 1:3] = imp_mean.fit_transform(data1.iloc[:, 1:3])
```

SimpleImputer(): Dùng để bổ sung+ thay thế các giá trị missing

missing_values=np.nan: các dữ liệu missing được thay thế

strategy='mean': sử dụng thay thế giá trị missing bằng mean (giá trị trung bình cộng)

imp_mean.fit_transform(): Phù hợp với dữ liệu imp_mean, sau đó biến đổi nó

data.iloc[]: Chọn dòng theo vị trí nguyên

# **Tính bằng tay, tính trung bình cộng rồi đặt vào các ô dữ liệu missing**




```python
data2 = pd.read_csv("categorical_data.csv")
```

```python
data2
```

Sử dụng fillna thay thế các dữ liệu missing trong cột

Thay thế dữ liệu missing ở cột age thành giá trị trung bình cộng của cột:

```python
data2['age'].fillna(data2['age'].mean(), inplace = True)
```

```python
data2
```

Thay thế dữ liệu missing ở cột salary thành giá trị trung bình cộng của cột:

```python
data2['salary'].fillna(data2['salary'].mean(), inplace = True)
```
